### Data Cleaning and Wrangling

```python
df_sales.head(1)
```

```python
# Merge with Customers
### Left join of Sales Orders & Customers table
df = df_sales.merge(
    df_customers,
    how = 'left',
    left_on = 'Customer Name Index',
    right_on = 'Customer Index'
)
```

```python
df.head(5)
```

```python
# Merge with Products

df = df.merge(
    df_products,
    how = 'left',
    left_on = 'Product Description Index',
    right_on = 'Index'
)
```

```python
df.head(5)
```

```python
# Merge with Regions

df = df.merge(
    df_regions,
    how = 'left',
    left_on = 'Delivery Region Index',
    right_on = 'id'
)
```

```python
df.head(5)
```

```python
# Merge with State Regions

df = df.merge(
    df_state_reg[["State Code", "Region"]],
    how = 'left',
    left_on = 'state_code',
    right_on = 'State Code'
)
```

```python
df.head(5)
```

```python
# Merge with Budgets

df = df.merge(
    df_budgets,
    how = 'left',
    on = 'Product Name'
)
```

```python
#Customer index
#Index
#id
#State Code

#Clean up redundant columns
cols_to_drop = ['Customer index', 'Index', 'id', 'State Code']
df = df.drop(columns = cols_to_drop, errors='ignore')

df.head(5)
```

```python
#Convert all columns to lower case for consistency and easier access
df.columns = df.columns.str.lower()

df.columns.values
```

```python
#Keep the important columns and delete the columns that we dont need

cols_to_keep = ['ordernumber', 'orderdate', 'customer names', 'channel', 'product name', 'order quantity', 'unit price',
       'line total', 'total unit cost', 'county', 'state_code', 'state', 'region', 'latitude', 'longitude', '2017 budgets']

# Subset the DataFrame to only these columns
df = df[cols_to_keep]
    
```

```python
df.head(5)
```

```python
#Rename the columns 

df = df.rename(columns={
    'ordernumber'      : 'order_number',   
    'orderdate'        : 'order_date',     
    'customer names'   : 'customer_name',  
    'product name'     : 'product_name',  
    'order quantity'   : 'quantity',     
    'unit price'       : 'unit_price',    
    'line total'       : 'revenue',       
    'total unit cost'  : 'cost',          
    'state_code'       : 'state',          
    'state'            : 'state_name',    
    'region'           : 'us_region',     
    'latitude'         : 'lat',            
    'longitude'        : 'lon',            
    '2017 budgets'     : 'budget'
})

df.head(1)
```

```python
# Blank out budgets for non-2017 orders
df.loc[df['order_date'].dt.year != 2017, 'budget'] = pd.NA

# Inspect
df[['order_date','product_name','revenue','budget']].head(10)
     
```

###Feature Engineering

```python
# Calculate total cost for each line item

df['total_cost'] = df['quantity'] * df['cost']
```

```python
# Calculate profit as revenue minus total_cost

df['profit'] = df['revenue'] - df['total_cost']
```

```python
# Calculate profit margin as a percentage

df['profit_margin_pct'] = (df['profit'] / df['revenue']) * 100
```
